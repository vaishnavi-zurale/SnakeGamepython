```python
import turtle
import random # ganrate to randam int(food) thats why its use 
import time # speed

delay=0.1
sc=0
hc=0
bodies=[]
#   *** creating a screen****
s=turtle.Screen()
s.title("Snake Game")
s.bgcolor("light blue")
s.setup(width=600,height=600) # size of screen

#  ***creating a head***

head=turtle.Turtle()
head.speed(0)
head.shape("square")
head.color("blue")
head.fillcolor("red")
head.goto(0,0)
head.penup()
head.direction=("stop")


#creating a color for a snake

food=turtle.Turtle()
food.penup()
food.speed(0)
food.shape("circle")
food.color("green")
food.fillcolor("light green")
food.hideturtle()
food.goto(150,200)
food.showturtle()

# creating a scorebord
sb=turtle.Turtle()
sb.penup()
sb.hideturtle()
sb.goto(-250,280)
sb.showturtle()
sb.write("score:0   |highst score:0") #to print a scoure on the screen for the 1st time

#creating function for moving in all direction
def moveUp():
 if head.direction!="Down":
     head.direction="Up"
def moveDown():
 if head.direction!="Up":
     head.direction="Down"
def moveLeft():
 if head.direction!="Right":
     head.direction="Left"
def moveRight():
 if head.direction!="Left":
     head.direction="Right"
def moveStop():
     head.direction="Stop"

def move():
    if head.direction=="Up":
     y=head.ycor()          #ycor or Y meance y axis
     head.sety(y+20)

    if head.direction=="Left":
     x=head.xcor()   #xcor or X, meance x axis
     head.setx(x-20)
    
    if head.direction=="Down":
     y=head.ycor()
     head.sety(y-20) 

    if head.direction=="Right":
     x=head.xcor()
     head.setx(x+20)

# event handling
s.listen()#s is screeen & listen() function is use for what screen do (kya kya kare )
s.onkey(moveUp,"Up") #onkey is useed on keybord key
s.onkey(moveDown,"Down")
s.onkey(moveLeft,"Left")
s.onkey(moveRight,"Right")
s.onkey(moveStop,"space")

#main loop
try:
   while True:
      s.update() #to update the screen

    # check ccollision with border
      if head.xcor()>290:
         head.setx(-290)
      if head.xcor()< -290:
         head.setx(290)
      if head.ycor()>290:
         head.sety(-290)
      if head.ycor()< -290:
         head.sety(290)

 #check collision with food

      if head.distance(food)<20:
         x=random.randint(-290,290)
         y=random.randint(-290,290)
         food.goto(x,y)

     #increase the body of snake
         body = turtle.Turtle()
         body.speed(0)
         body.penup()
         body.shape("square")
         body.color("red")
         bodies.append(body) # append the new body in list

         sc=sc+100
         delay=delay-0.001

         if sc>hc:
            hc=sc #update highest score
         sb.clear()
         sb.write("Score:{}    |   Highest Score:{}".format(sc, hc))

     #***move snake body
     
      for i in range(len(bodies)-1,0,-1):
        x=bodies[i-1].xcor()
        y=bodies[i-1].ycor()
        bodies[i].goto(x,y)

      if len(bodies)>0:
         x=head.xcor()
         y=head.ycor()
         bodies[0].goto(x,y)
      move()
# check collision with snake body--
      for body in bodies:
         if body.distance(head)<20:
             time.sleep(1)
             print("game over")
             head.goto(0,0)
             head.direction="stop"
             

         #hide body'''
             for body in bodies:
                body.hideturtle()
                bodies.clear()
                sc=0
                delay=0.1
                sb.clear()
                sb.write("Score:{}  | Highest Score:{}".format(sc, hc))
                bodies.clear()
      time.sleep(delay)
      
except turtle.Terminator: 
    pass
#s.mainloop()
```
